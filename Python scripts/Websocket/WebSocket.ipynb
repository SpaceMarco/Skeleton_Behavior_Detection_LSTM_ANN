```python
import asyncio
import websockets
import zmq
import numpy as np
from keras.models import load_model

LSTM_model = load_model('lstm_model.h5')
Ann_model = load_model('ann_model.h5')





async def rec(websocket):
    while True:
        m = await websocket.recv()

        test_arr = m.replace(")(", ',').replace(')', '').replace('(', '').split(',')

        test_arr = np.array(test_arr)
        test_arr = np.float_(test_arr)


        yhat = LSTM_model.predict(test_arr.reshape(1,69))
        ann_out = Ann_model.predict(yhat)

        y_pred_classes = [np.argmax(i) for i in ann_out]

        print("\n-------------> "+ str(y_pred_classes[0]))

        await websocket.send(str(y_pred_classes[0]))
        # print(test_arr)

    # async for message in websocket:
        

async def main():
    async with websockets.serve(rec, "localhost", 5555):
        await asyncio.Future()  # run forever

await main()
# await rec()
```
